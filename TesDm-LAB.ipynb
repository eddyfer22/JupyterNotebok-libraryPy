# Plotly
Es una librería muy fundamental para la visualización y compresión de datos que brinda objetos gráficos fáciles de usar y representar casos de estudio científicos y estadísticos.

```python
import plotly.express as px
import plotly.graph_objects as go
```

### Scatter
Se realiza una grafica de puntos utilizando dos columnas del dataframe

```python
fig = px.scatter(df, x="total_tracks", y="t_dur0")
fig.show()
```

### Area 
Se genera el grafico con la funcion area en el que se muestra el total de tracks y duracion de las canciones

```python
fig2 = px.area(df, x="total_tracks", y="t_dur0", title='Tracks y duracion')
fig2.show()
```

### Line
Se realiza un grafico en los ejes de x e y con la funcion Line

```python
fig2 = px.line(df, x="name", y="t_dur0", title='Tracks y duracion')
fig2.show()
```

# Seaborn
Es una librería importante para la gráfica, visualización de datos y análisis exploratorios de los datos en el que proporciona facilidades en encontrar tendencias en los datos. 

```python
import seaborn as sns
```

### Relplot
Con la gestionan la figura y el conjunto de ejes que las contiene, muestrando la relación entre dos variables.

```python
sns.set_theme()
sns.relplot(
    data=df,
    x="t_dur0", y="total_tracks")
```

### Catplot
Se realiza un grafico que proporciona acceso a varias funciones a nivel de ejes que muestran la relación entre una variable numérica

```python
sns.catplot(
    data=df,
    x="total_tracks", y="t_dur0")
```

### Displot
Se crea un grafico que proporciona acceso a varios enfoques para visualizar la distribución univariante o bivariada de datos incluidos subconjuntos de datos. 

```python
sns.displot(
    data=df,
    x="total_tracks", y="t_dur0")
```
